```python
from selenium.common.exceptions import NoSuchElementException
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import time
```

```python
driver = webdriver.Chrome()
driver.get("https://twitter.com/")
```

```python

n = 0
indexnum = 1
point_num = 5   
df = pd.DataFrame(columns = ['username','name','content','time','likes','reply','retweet'])

while True:

    target = driver.find_elements(By.XPATH,"//div[@data-testid='cellInnerDiv']")

    for j in target:

        try:
            username = j.find_element(By.CSS_SELECTOR,"div[class='css-1dbjc4n r-18u37iz r-1wbh5a2 r-13hce6t'] div[class='css-1dbjc4n r-1wbh5a2 r-dnmrzs'] span[class='css-901oao css-16my406 r-poiln3 r-bcqeeo r-qvutc0']")
            df.loc[indexnum,'username']= username.text
            
        except NoSuchElementException:

            df.loc[indexnum,'username']= "None"  

        try:
            name = j.find_element(By.CSS_SELECTOR,'div[class="css-901oao r-1awozwy r-18jsvk2 r-6koalj r-37j5jr r-a023e6 r-b88u0q r-rjixqe r-bcqeeo r-1udh08x r-3s2u2q r-qvutc0"] > span >span')
           
            df.loc[indexnum,'name']= name.text
            
        except NoSuchElementException:

            df.loc[indexnum,'name']= "None"

        try:

            content = j.find_element(By.CSS_SELECTOR,"div[data-testid='tweetText'] ")
            df.loc[indexnum,'content']= content.text
            
        except NoSuchElementException:

            df.loc[indexnum,'content']= "None"
        try:

            timee = j.find_element(By.CSS_SELECTOR,'div[class="css-1dbjc4n r-18u37iz r-1q142lx"] a[role="link"] time').get_attribute('datetime')
            df.loc[indexnum,'time']= timee
            
        except NoSuchElementException:

            df.loc[indexnum,'time']= "None"
        try:

            likes = j.find_element(By.CSS_SELECTOR,'div[class="css-1dbjc4n r-xoduu5 r-1udh08x"] span[data-testid="app-text-transition-container"] span[class="css-901oao css-16my406 r-poiln3 r-n6v787 r-1cwl3u0 r-1k6nrdp r-1e081e0 r-qvutc0"] span[class="css-901oao css-16my406 r-poiln3 r-bcqeeo r-qvutc0"]')
            df.loc[indexnum,'likes']= likes.text

        except NoSuchElementException:

            df.loc[indexnum,'likes']= "None"

        try:

            reply = j.find_element(By.CSS_SELECTOR,'span[data-testid="app-text-transition-container"] span[class="css-901oao css-16my406 r-poiln3 r-n6v787 r-1cwl3u0 r-1k6nrdp r-1e081e0 r-qvutc0"] span[class="css-901oao css-16my406 r-poiln3 r-bcqeeo r-qvutc0"]')
            df.loc[indexnum,'reply']= reply.text

        except NoSuchElementException:

            df.loc[indexnum,'reply']= "None"
        try:

            retweet = j.find_element(By.CSS_SELECTOR,'span[data-testid="app-text-transition-container"] span[class="css-901oao css-16my406 r-poiln3 r-n6v787 r-1cwl3u0 r-1k6nrdp r-1e081e0 r-qvutc0"] span[class="css-901oao css-16my406 r-poiln3 r-bcqeeo r-qvutc0"]')
            df.loc[indexnum,'retweet']= retweet.text

        except NoSuchElementException:

            df.loc[indexnum,'retweet']= "None"

        indexnum += 1
        n+=1
        time.sleep(1)

    if n > point_num:
        point_num += 5
        driver.execute_script('window.scrollTo(0,document.body.scrollHeight)')
        time.sleep(5)
        print('scrolling')

        if n >= 100:
            break

        else:
            continue
```
